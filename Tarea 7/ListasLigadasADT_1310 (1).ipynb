```python
class Nodo:
  def __init__(self, value, siguiente = None):
    self.data = value
    self.next = siguiente
```

```python
class LinkedList:
  def __init__(self):
    self.head = None

  def is_empty(self):
    return self.head == None

  def get_tail(self):
    if self.is_empty():
      return None
    else:
      curr_node = self.head
      while curr_node.next != None:
        curr_node = curr_node.next
      return curr_node 

  def append(self, value):
    if self.head == None:
      self.head = Nodo(value)
    else: 
      curr_node = self.head
      while curr_node.next != None:
        curr_node = curr_node.next
      curr_node.next = Nodo(value)

  def prepend(self, value):
    if self.head == None:
      print('Está vacio, no se puede hacer prepend')
    else:
      aux = self.head
      self.head = Nodo(value, aux)

  def transversal(self):
    curr_node = self.head
    while curr_node != None:
      print(f'| {curr_node.data} |-->', end='')
      curr_node = curr_node.next
    print('')

  def remove(self, value):
    if self.head == None:
      print('Lista vacia no se puede remover nada')
    else:
      curr_node = self.head
      aux = None
      while curr_node.data != value:
        aux = curr_node
        curr_node = curr_node.next
      if curr_node != None:
        aux.next = curr_node.next
        curr_node = None

  def pop(self):
        if self.head == None :
            print("Lista vacía, no es posible regresar el ultimo valor.")
        else :
            curr_node=self.head
            aux = None
            while curr_node.next != None:
                aux= curr_node
                curr_node=curr_node.next
            aux.next = None
            return curr_node.data

  def pop_first(self):
        if self.head == None:
            print("Lista vacía, imposible eliminar el primer valor.")
        else:
            curr_node = self.head
            self.head = curr_node.next
            return curr_node.data
```

```python
lista = LinkedList()
if lista.is_empty():
  print('La lista está vacia')
else:
  print('La lista no está vacia')

lista.append(10)
lista.append(20)
lista.append(30)
lista.append(40)
lista.append(50)

print("Tail")
nodo = lista.get_tail()
if nodo != None:
  print(nodo.data)
else:
  print('La lista está vacia')

print("Transversal")
lista.transversal()

print("Remove")
lista.remove(20)
lista.transversal()

print("Prepend")
lista.prepend(0)
lista.transversal()

print("Pop")
print(f'Valor eliminado : {lista.pop()}')
lista.transversal()

print("Pop first")
print(f'Valor eliminado : {lista.pop_first()}')
lista.transversal()
```
